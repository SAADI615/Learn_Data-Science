```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]

y = [5,6,7]

yes = plt.plot(y,x, label = 'yx', color = 'red', linewidth = 1, marker = '.', markersize = 10, markeredgecolor = 'blue', linestyle = '--')

no = plt.plot(x,y, 'g^--', label = 'fresh')

plt.title('My frist graph', fontdict = {'fontname' : 'Arial', 'fontsize': 14})

plt.xlabel('X axis', fontdict = {'fontsize': 14})

plt.ylabel('Y axis', fontdict = {'fontsize': 14})

plt.xticks([])

plt.yticks([])

plt.legend()

plt.show()
```

```python
x2 = np.arange(0,65)

plt.figure(figsize = (4,2), dpi = 100)

plt.plot(x2[:22],x2[:22]**2, label = 'Present')
plt.plot(x2[21:65],x2[21:65]**2, 'r-', label = 'In_future')
plt.yticks([])
plt.title('Insallah')
plt.xlabel('My age')
plt.ylabel('My growth')
plt.legend()
plt.show()
plt.savefig('future.png', dpi = 300)
```

```python
labels = ['A', 'B', 'C']
values = [20,30,50]
plt.figure(figsize=(5,3), dpi = 100)

bars = plt.bar(labels,values, label = 'bar')

patterns = ['/', 'O', '*','-']

for i in bars:
    i.set_hatch(patterns.pop(0))

plt.title('The bar')

plt.legend()
plt.show()
```
