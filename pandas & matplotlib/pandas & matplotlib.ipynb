```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('Display.max_columns', 90)
```

```python
gas_df = pd.read_csv('Data/gas_prices.csv')
fifa_df = pd.read_csv('Data/fifa_data.csv')
```

```python
plt.figure(figsize=(8,5), dpi=100)

# plt.plot(gas_df.Year, gas_df.USA, 'b.-', label = 'USA')
# plt.plot(gas_df.Year, gas_df.Canada, 'g.-', label = 'Canada')
# plt.plot(gas_df.Year, gas_df['South Korea'], 'r.-', label = 'South Korea')
# plt.plot(gas_df.Year, gas_df.Australia, 'y.-', label = 'Australia')

countries = ['Canada', 'UK']

for column in gas_df:
    if column != 'Year':
        plt.plot(gas_df.Year, gas_df[column] ,label = column, marker = '.')

plt.title('Gas Price($)', fontdict={'fontweight':'bold', 'fontsize':22})
plt.xlabel('Year')
plt.ylabel('USD')
plt.xticks(gas_df.Year[::2].tolist()+[2011])

plt.legend()
plt.savefig('Gas_price_figure.png', dpi = 300)
plt.show()
```

```python
x_bars = [40,50,60,70,80,90,100]
plt.hist(fifa_df.Overall, color='lime')

plt.ylabel('Numbers of player')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.xticks(x_bars)
plt.show()
```

```python
fifa_df['Preferred Foot'].value_counts()
```

```python
left = fifa_df.loc[fifa_df['Preferred Foot'] == 'Left'].count()[0]
right = fifa_df.loc[fifa_df['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']

plt.title('Foot Preference of FIFA Players')

plt.pie([left,right], labels = labels, colors = ['gray','lime'], autopct='%.2f %%')
plt.show()
```

```python
fifa_df.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa_df.Weight]
light = fifa_df.loc[fifa_df.Weight < 125].count()[0]
light_medium = fifa_df.loc[(fifa_df.Weight >= 125) & (fifa_df.Weight < 150)].count()[0]
medium_heavy = fifa_df.loc[(fifa_df.Weight >= 150) & (fifa_df.Weight < 175)].count()[0]
heavy = fifa_df.loc[fifa_df.Weight >= 175].count()[0]

plt.style.use('ggplot')
explode = (.5,0,0,0)

plt.pie([light, light_medium, medium_heavy, heavy], labels = ['light','light medium','medium heavy','heavy'], autopct='%.2f %%', pctdistance = 0.9, explode=explode)


plt.title('FIFA players weight')
plt.show()
```

```python
barcelona = fifa_df.loc[fifa_df['Club'] == 'FC Barcelona']['Overall']
madrid = fifa_df.loc[fifa_df['Club'] == 'Real Madrid']['Overall']
revs = fifa_df.loc[fifa_df['Club'] == 'New England Revolution']['Overall']

plt.figure(figsize=(6,8))
labels = ['FC Barcelona', 'Real Madrid','New England Revolution']

boxes =plt.boxplot([barcelona,madrid,revs], labels = labels, patch_artist=True, notch=True, vert=True)

color=['red','green','blue']
for box in boxes ['boxes']:
    box.set(color='red')

    box.set(facecolor='blue')

plt.title('Proffetonal Football Team Comparison')
plt.ylabel('Overall rating')
plt.show()
```
